```python
import numpy as np
import time
```

```python
def random_fn():
    time.sleep(0.25)
    print(f"time.sleep(0.25)")
    
    a = np.random.randint(0,1000, (1, 1000))
    print(f"a = np.random.randint(0,1000, (1, 1000))")
    print(f"np.shape(a) = {np.shape(a)}")
    
    b = np.tile(a, (1000, 1))
    print("b = np.tile(a, (1000, 1))")
    print(f"np.shape(b) = {np.shape(b)}")
    
    c = a*b
    print('c = a*b')
    print(f"np.shape(c) = {np.shape(c)}")
    
    d = np.random.randint(0, 1000, (np.shape(c)))
    print('d = np.random.randint(0, 1000, (np.shape(c)))')
    print(f"np.shape(d) = {np.shape(d)}")
    
    e = c*d
    print('e = c*d')
    print(f"np.shape(e) = {np.shape(e)}")
    
    time.sleep(0.25)
    print(f"time.sleep(0.25)")
```

```python
logging_interval=0.00001
print(f"logging_interval = {logging_interval} times per second")
```

```python
monitor = ResourceMonitor(logging_interval=logging_interval)
monitor.start_logging()

random_fn()

monitor.stop_logging()
```
